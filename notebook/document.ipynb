```python
## Create a simple txt file
import os
os.makedirs("../data/text_files",exist_ok=True)
```

```python
sample_texts={
    "../data/text_files/python_intro.txt":"""Python Programming Introduction

Python is a high-level, interpreted programming language known for its simplicity and readability.
Created by Guido van Rossum and first released in 1991, Python has become one of the most popular
programming languages in the world.

Key Features:
- Easy to learn and use
- Extensive standard library
- Cross-platform compatibility
- Strong community support

Python is widely used in web development, data science, artificial intelligence, and automation.""",
    
    "../data/text_files/machine_learning.txt": """Machine Learning Basics

Machine learning is a subset of artificial intelligence that enables systems to learn and improve
from experience without being explicitly programmed. It focuses on developing computer programs
that can access data and use it to learn for themselves.

Types of Machine Learning:
1. Supervised Learning: Learning with labeled data
2. Unsupervised Learning: Finding patterns in unlabeled data
3. Reinforcement Learning: Learning through rewards and penalties

Applications include image recognition, speech processing, and recommendation systems
    
    
    """

}

# Loop through each filepath and its corresponding text content in the dictionary
for filepath, content in sample_texts.items():
    # Open the file in write mode ('w') with UTF-8 encoding to handle any special characters
    with open(filepath, 'w', encoding="utf-8") as f:
        # Write the text content into the file
        f.write(content)

# Print a success message once all sample files are successfully written
print("✅ Sample text files created!")
```

```python
# Import TextLoader from the community package, which is designed to read text files and convert them into LangChain Document objects
from langchain_community.document_loaders import TextLoader

# Initialize the loader with the target file path and specify UTF-8 encoding to prevent formatting/special character errors
loader = TextLoader("../data/text_files/python_intro.txt", encoding="utf-8")

# Load the file content into memory. This reads the file and packages it into a list of LangChain Document objects (retaining the content and source metadata)
document = loader.load()

# Print the loaded document list to verify that the text has been successfully read and structured
print(document)
```

```python
# Import DirectoryLoader to read multiple files from a directory at once
from langchain_community.document_loaders import DirectoryLoader

# Set up the DirectoryLoader to scan a folder and load all matching files
dir_loader = DirectoryLoader(
    "../data/text_files",                   # The target directory path containing our files
    glob="**/*.txt",                         # Search pattern to recursively match all text files (.txt)
    loader_cls=TextLoader,                   # Instructs DirectoryLoader to use TextLoader to process each file
    loader_kwargs={'encoding': 'utf-8'},     # Arguments passed to the TextLoader (ensures files open in UTF-8)
    show_progress=False                      # Set to True if you want a progress bar when loading many files
)

# Load all matching text files in the directory as a list of LangChain Document objects
documents = dir_loader.load()

# Display the loaded documents list to inspect the contents and metadata of all loaded files
documents
```

```python
# Import PyPDFLoader and PyMuPDFLoader to load and parse PDF files
from langchain_community.document_loaders import PyPDFLoader, PyMuPDFLoader

# Configure the DirectoryLoader to look inside the pdf folder
dir_loader = DirectoryLoader(
    "../data/pdf",                          # The target directory path containing your PDF files
    glob="**/*.pdf",                        # Search pattern to recursively match all PDF files (.pdf)
    loader_cls=PyMuPDFLoader,                # Use PyMuPDFLoader to process and extract text from each PDF file
    show_progress=False                     # Disable the loading progress bar
)

# Scan the folder, parse each PDF file found, and convert them into LangChain Document objects
pdf_documents = dir_loader.load()

# Display the extracted documents along with their content and metadata (like page numbers)
pdf_documents
```
